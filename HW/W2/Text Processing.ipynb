```python
import re
text = "555-1239Dr. Bernard Lander(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Sco\tt555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"
```

### Part 1

```python
res = re.findall(r'\d+', text)
print (res)
```

### Part 2

```python
Title = r"(?:[A-Z][a-z]*\.\s*)*"
FirstName = r"[A-Z][a-z]+,*\s+"
MiddleName = r"(?:[A-Z][a-z]*\.?\s*)?"
LastName = r"[A-Z][a-z]+"

res = re.findall(Title + FirstName + MiddleName + LastName, text)
print (res)
```

### Part 3

```python
res = re.findall(Title + FirstName + MiddleName + LastName, text)
updatedRes = [i.replace (',' , '')for i in res ]
print (updatedRes)
```

### Part 4

```python
isTitle = []
isTitleBool = []
tempTitle = []

for i in range (0,len(updatedRes)):
  title = re.findall('[a-zA-Z]+\.', updatedRes[i])
  tempTitle.append(title) 

for i in range (0,len(tempTitle)):
  
  if not tempTitle[i]:
    isTitle.append (f"{i}: Does Not Have Title                        and Full Name: {updatedRes[i]}")
    isTitleBool.append (f"{i}: {False}")
  else:
    isTitle.append (f"{i}: Has Title,          named: {tempTitle[i][0]} and Full Name: {updatedRes[i]}")
    isTitleBool.append (f"{i}: {True}")


print (isTitleBool, "\n")
isTitle
```

### Part 5

```python
isMiddle = []
isMiddleBool = []
for i in range (0,len(updatedRes)):

  x = updatedRes[i].split()
  y = len(x)
  if isTitleBool[i]:

    if y>3:
      isMiddle.append (f"{i}: Has Middle/Second Name,          named: {x[2]} and Full Name: {updatedRes[i]}")
      isMiddleBool.append (f"{i}: {True}")
    else:
      isMiddle.append (f"{i}: No Middle/Second Name,           and Full Name: {updatedRes[i]}")
      isMiddleBool.append (f"{i}: {False}")
  else:

    if y>2:
        isMiddle.append (f"{i}: Has Middle/Second Name,          named: {x[2]} and Full Name: {updatedRes[i]}")
        isMiddleBool.append (f"{i}: {True}")

    else:
        isMiddle.append (f"{i}: No Middle/Second Name,           and Full Name: {updatedRes[i]}")
        isMiddleBool.append (f"{i}: {False}")

print (isMiddle)
print (isMiddleBool)
```

### Part 6


    This fails because:
      '<.+>' means  start with < then any character one or more occurences then end with >. 
      but regular expression takes > as a character and ends in the very end. So that it includes >. Eventually prints the whole string.

```python
textt = "<title>+++BREAKING NEWS+++</title>"
res = re.findall('<.+>', textt)
print(res)
```

#### Corrected

```python
textt = "<title>+++BREAKING NEWS+++</title>"
res = re.findall('<[/]?.+?>' , textt)
print(res)
```
