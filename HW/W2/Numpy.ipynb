### Problem 6

```python
a = np.array([12, 5, 7, 15, 3, 1, 8])
b = np.array([14, 6, 3, 11, 19, 12, 5])
temp = []
for i in range (0,len(a)):
    if a[i] not in b:
        temp.append(a[i]) 
c= np.array(temp)
print (c)
```
