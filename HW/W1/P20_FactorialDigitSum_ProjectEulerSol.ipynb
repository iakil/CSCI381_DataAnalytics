## Solution:

```python
class Solution:
    def sumOfFact(self,num):
       #num = int (num)
        factoriall =1
        for i in range (1,num+1):
           factoriall = factoriall * i

        factoriall = str (factoriall)
        total = 0
        for i in range (0,len(factoriall)):
            total = total + int (factoriall[i])
        return total
```
