```python
class Solution: 
   def longestChain(self):
      highest = 0
      result = []
     
      for i in range (1,1000001):
         tracking = s1.collatz(i)[1]
         if tracking > highest:
            result = []
            highest = tracking
            result = [i, highest]

      return result

   def collatz (self, num): 
      my_collatz = [num]
      count = 0
      while num != 1:
         
         if num % 2 == 0:
            num = num//2
            my_collatz.append(num)
            count = count +1
         else:
            num = num*3+1 
            my_collatz.append(num)
            count = count +1
         
      return [my_collatz,count]
```
